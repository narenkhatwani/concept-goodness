```python
# List of TUIs to filter out
tuis_to_filter = [
    'T109', 'T121', 'T123', 'T195', 'T103', 'T104', 'T200',
    'T011', 'T008', 'T012', 'T013', 'T015', 'T002', 'T014',
    'T074', 'T075', 'T196', 'T028', 'T204'
]

# Drop rows where 'TUI' is in the list of TUIs to filter out
filtered_df = merged_df[~merged_df['TUI'].isin(tuis_to_filter)]
```

## Deleting more chemical names

If the name string contains digits like here 1,2-Dipalmitoyl-Glycerophosphocholine then we are deleting that row as they correspond to chemical names/prescriptions

```python
filtered_df = filtered_df[~filtered_df['STR'].str.contains(r'\d', na=False)]
```

```python
filtered_df
```

# Similar CUIs deletion

- Normalize the strings by removing punctuation and converting to lowercase.
- Sort the words within each string to identify redundant variations.
- Group by CUI and retain only the unique variations based on the normalized and sorted strings.

```python
# Function to normalize strings
def normalize_string(s):
    if isinstance(s, str):
        s = re.sub(r'[^\w\s]', '', s)  # Remove punctuation
        s = s.lower()  # Convert to lowercase
        words = s.split()  # Split into words
        words.sort()  # Sort words alphabetically
        return ' '.join(words)
    return s

# Apply normalization
filtered_df['normalized_STR'] = filtered_df['STR'].apply(normalize_string)

# Drop duplicates based on CUI and normalized_STR, keeping the first occurrence
unique_df = filtered_df.drop_duplicates(subset=['CUI', 'normalized_STR'])

# Drop the normalized_STR column
unique_df = unique_df.drop(columns=['normalized_STR'])

```

```python
# Function to get 100 random samples for each semantic type
def sample_semantic_types(df, sample_size=100):
    return df.groupby('STY').apply(lambda x: x.sample(min(len(x), sample_size))).reset_index(drop=True)

# Get 100 random samples for each semantic type
sampled_df = sample_semantic_types(random_sampling, sample_size=500)
```
